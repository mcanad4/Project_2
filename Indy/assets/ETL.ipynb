```python
import os
import pandas as pd
from sqlalchemy import create_engine
from config import username, password, dbhost, dbport, dbname 
```

```python
connection_string = f'postgresql://{username}:{password}@{dbhost}:{dbport}/{dbname}'

engine = create_engine(connection_string)
conn = engine.connect()
```

```python
bus_stops = pd.read_sql('select * from bus_stops', con=engine)
bus_stops
```

```python
# Export file to csv
bus_path = os.path.join('.', 'assets', 'Clean', 'bus_clean.csv')
bus_stops.to_csv(bus_path, header=True, index=False)
```

```python
covid_marion_county = pd.read_sql('select * from covid_marion_county', con=engine)
covid_marion_county
```

```python
covid_marion_county['patient_count'] = covid_marion_county['patient_count'].astype(str)
```

```python
covid_marion_county['population'] = covid_marion_county['population'].astype(str)
```

```python
covid_marion_county['percentage'] = covid_marion_county['percentage'].astype(str)
```

```python
covid_marion_county.replace({'patient_count':'nan'},{'patient_count':'0'},inplace=True)
covid_marion_county.replace({'population':'nan'},{'population':'0'},inplace=True)
covid_marion_county.replace({'percentage':'nan'},{'percentage':'0'},inplace=True)
covid_marion_county
```

```python
# Export file to csv
covid_path = os.path.join('.', 'assets', 'Clean', 'covid_clean.csv')
covid_marion_county.to_csv(covid_path , header=True, index=False)
```
